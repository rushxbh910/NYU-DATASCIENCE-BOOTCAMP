1. Define two custom numpy arrays, say A and B. Generate two new numpy
arrays by stacking A and B vertically and horizontally.

```python
import numpy as np
a = np.array([[1, 2, 3],[4,5,6]]) #2-D Array
print("a: ",a)
b=np.array([[7, 8, 9],[10,11,12]]) #2-D Array
print("b: ",b)
print("Shape of a: ",a.shape)
print("Shape of b: ",b.shape)

c=np.vstack((a,b))
print("Row Stack: ")
print(c)

d=np.hstack((a,b))
print("Col Stack: ")
print(d)
```

2. Find common elements between A and B. [Hint : Intersection of two sets]

```python
a = np.array([1, 2, 3,4,5,6]) #2-D Array
print("a: ",a)
b=np.array([5,6,7,8,9,10]) #2-D Array
print("b: ",b)
n= len(a)
m= len(b)
for i in range(n):
    for j in range(m):
        if a[i]==b[j]:
            print("Common Elemnt is at ",i+1,"index in a and ",j+1,"index in b ",a[i],",",b[j])
```

Using built function intersect1d()

```python
import numpy as np
a= np.array([1,2,3,4,5,6])
b= np.array([5,6,7,8,9])
cmn_ele=np.intersect1d(a,b)

print("common elements ", cmn_ele)
```

3. Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
import numpy as np
a= np.array([1,2,3,4,5,6,10,15,20,50,65,80,100])

i=np.where((a>4) & (a<11)) #This will find all indices at which the value of a is b/w 5 and 10 (inclusive)
print(i) #Prints indices
print(a[i]) #Prints values at indices
```

5. From df filter the 'Manufacturer', 'Model' and 'Type' for every
20th row starting from 1st (row 0).

```python
import pandas as pd
url='https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df=pd.read_csv(url)
filter = df.iloc[::20,[df.columns.get_loc('Manufacturer'), df.columns.get_loc('Model'), df.columns.get_loc('Type')]]
print(filter)


```

6. Replace missing values in Min.Price and Max.Price columns with
their respective mean.

DF before replacing min and max:

```python
import pandas as pd
url='https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df=pd.read_csv(url)
df.head(10)
```

DF after replacing minprice and maxprice with respective avg:

```python
import pandas as pd
url='https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df=pd.read_csv(url)
avg_min_price=df['Min.Price'].mean()
avg_max_price=df['Max.Price'].mean()
df['Min.Price'].fillna(avg_min_price,inplace=True)
df['Max.Price'].fillna(avg_max_price,inplace=True)
df.head(10)
```

7. How to get the rows of a dataframe with row sum > 100?

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print("df before filter: \n",df)
filter=df[df.sum(axis=1)>100]
print("\n df after filter: \n",filter)
```
